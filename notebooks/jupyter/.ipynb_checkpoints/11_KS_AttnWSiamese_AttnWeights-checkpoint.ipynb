### Get attention weights

```python
attn_dict  = {'fold_'+str(fold):ReaderWriter.read_data(os.path.join(up_dir, tr_val_dir + '/test/fold_' + str(fold), 'sampleid_fattnw_map_test.pkl')) for fold in range(5)}
```

```python
def merge_dicts(*dict_args):
    """
    Given any number of dictionaries, shallow copy and merge into a new dict,
    precedence goes to key value pairs in latter dictionaries.
    """
    result = {}
    for dictionary in dict_args:
        result.update(dictionary)
    return result
```

```python
attn_dict_Xa = merge_dicts(*[attn_dict['fold_' + str(fold)]["X_a"] for fold in range(num_folds)])
attn_dict_Xb = merge_dicts(*[attn_dict['fold_' + str(fold)]["X_b"] for fold in range(num_folds)])
```

### Split per prediction case

```python
prediction_cases = [['TP', 1, 1], ['FP', 0, 1], ['TN', 0, 0], ['FN', 1, 0]]
sum(prediction_cases, [])[::3]
```

```python
case_IDs = {case[0]:predictionsTest[(predictionsTest.true_class == case[1]) & (predictionsTest.pred_class == case[2])].id.tolist() for case in prediction_cases}
case_IDs
```

```python
# (t)ensor of drug (a) and drug (b) attention weights
t_ab_perCase = {case:torch.stack([torch.mean(torch.stack([attn_dict_Xa[i], attn_dict_Xb[i]]),0) for i in case_IDs[case]]) for case in list(case_IDs)}
t_ab_perCase
```

```python
# (t)ensor of drug (a) and drug (b) attention weights
t_ab = torch.stack([torch.mean(torch.stack([attn_dict_Xa[i], attn_dict_Xb[i]]),0) for i in list(predictionsTest["id"])])
t_ab.shape
```

```python
# Mean values per Similarity type
pymean_ab = torch.mean(t_ab, 0).tolist()
pymean_ab
```

### Plot weight distributions

```python
plt.figure(figsize=(15,25))
fig, axes = plt.subplots(ncols=3, nrows=3)
plt.subplots_adjust(top=1.4, bottom=0.1)

len_sim = len(similarity_types_GIP)

for i, axi in zip(range(len_sim), axes.flat):
    data = t_ab[:,i]
    axi.axvline((1/len_sim), color='red')
    ax = sns.distplot(data, hist=True, ax=axi)
    ax.set_title(similarity_types_GIP[i])
plt.show()
```

### Calculate ranks

```python
from scipy.stats import rankdata

reverseRank = True
totalRank_allCases = []
p_ab_perCase_ranked_allCases = []

for case in list(case_IDs):
    p_ab_perCase_List = t_ab_perCase[case].tolist()

    if (reverseRank):
        p_ab_perCase_ranked = [rankdata([-1 * i for i in x]).tolist() for x in p_ab_perCase_List]
        p_ab_perCase_ranked_allCases.append(p_ab_perCase_ranked)
    else:
        p_ab_perCase_ranked = [rankdata(x).tolist() for x in p_ab_perCase_List]
    totalRank_perCase = rankdata(torch.sum(torch.tensor(p_ab_perCase_ranked),0).tolist()).tolist()
    totalRank_allCases.append(totalRank_perCase)
```

```python
p_ab_tp_tn_mean = torch.mean(torch.tensor(p_ab_perCase_ranked_allCases[0] + p_ab_perCase_ranked_allCases[2]),0).tolist()
p_ab_tp_tn_mean
```

```python
totalRank_tp_tn = rankdata(p_ab_tp_tn_mean).tolist()
totalRank_tp_tn
```

```python
df_attn_scores = pd.DataFrame({'similarities':similarity_types_GIP, 'absolute':pymean_ab, 'ranked':totalRank_tp_tn})
df_attn_scores
```

```python
from matplotlib import pyplot
import matplotlib.patches as mpatches
from matplotlib.colors import ListedColormap

plt.close('all')
fig3 = plt.figure(constrained_layout=False)
gs = fig3.add_gridspec(5, 9, hspace=0.1, top=1.1)

f3_ax0 = fig3.add_subplot(gs[0, :])
f3_ax1 = fig3.add_subplot(gs[1, :])
f3_ax2 = fig3.add_subplot(gs[2, :])
f3_ax3 = fig3.add_subplot(gs[3, :])
f3_ax4 = fig3.add_subplot(gs[4, :])

red_patch = mpatches.Patch(color='red', label='Biological')
blue_patch = mpatches.Patch(color='blue', label='Phenotypic')
green_patch = mpatches.Patch(color='green', label='Chemical')

fontsize = 15
font_scale = 1.33

list1_9 = list(range(10))[1:]

if (DSdataset_name == 'DS1'):
    letters_space = 9.25
if (DSdataset_name == 'DS3'):
    letters_space = 8.25


def make_ranked_plot(ax, title, xlabel, ylabel, dict_colors, fontsize=fontsize, ytickfont=fontsize):
    ax.set_title(title, fontsize=fontsize)
    ax.set_xlabel(xlabel, fontsize=int(fontsize/3))
    ax.set_ylabel(ylabel, fontsize=int(fontsize/3))
    ax.set_yticklabels(ax.yaxis.get_ticklabels(), fontsize=ytickfont)
    return ax

def make_absolute_plot(ax, title, xlabel, ylabel, dict_colors, fontsize=fontsize):
    ax.set_title(title, fontsize=fontsize)
    ax.set_xlabel(xlabel, fontsize=int(fontsize/3))
    ax.set_ylabel(ylabel, fontsize=int(fontsize/3))
    [t.set_color(dict_colors[t.get_text()]) for t in ax.xaxis.get_ticklabels()]
    legend1 = pyplot.legend(handles=[blue_patch, red_patch, green_patch], 
                            bbox_to_anchor=(-0.19, -0.2), 
                            loc=2, fontsize='x-small',
                            facecolor='white', framealpha=0)
    ax.add_artist(legend1)
    return ax

# Subplot A
sns.set(font_scale=1.0)
ax_abs_fake = make_ranked_plot(ax = sns.heatmap(p_fake, 
                 xticklabels=[], 
                 yticklabels=['DrugPair 1', 'DrugPair 2', 'DrugPair 3', '...', 'DrugPair K'], 
                 robust=True, 
                 annot=True,fmt=".2", 
                 square=False,
                 cbar=False,
                 ax=f3_ax0
                                ), 
                title='Attention weights (example)', xlabel='', ylabel='', dict_colors=dict_sims_colors, ytickfont=12)
ax_abs_fake.text(letters_space, 3,'A', fontsize=25)

# Subplot B
sns.set(font_scale=1.5)
ax_abs = make_ranked_plot(ax = sns.heatmap([df_attn_scores['absolute']], 
                 xticklabels=[], 
                 yticklabels=['Avg\nWeight\n(actual)'], 
                 robust=True, 
                 annot=True,fmt=".2g", 
                 square=True,
                 cbar=False,
                 ax=f3_ax1
                                 ),
         title='', xlabel='', ylabel='', dict_colors=dict_sims_colors)
ax_abs.text(letters_space, 0.6,'B', fontsize=25)

# Subplot C
sns.set(font_scale=1.0)
ax_rank_fake = make_ranked_plot(ax = sns.heatmap(p_fake_ranked, 
                 xticklabels=[], 
                 yticklabels=['DrugPair 1', 'DrugPair 2', 'DrugPair 3', '...', 'DrugPair K'], 
                 robust=True, 
                 annot=True,fmt=".1g", 
                 square=False,                 
                 cbar=False,
                 cmap=sns.cm.rocket_r,
                 ax=f3_ax2
                                ), 
                title='Attention weights - Ranked (example)', xlabel='', ylabel='', dict_colors=dict_sims_colors, ytickfont=12)
ax_rank_fake.text(letters_space, 3,'C', fontsize=25)

# Subplot D
sns.set(font_scale=1.5)
ax_rank_avg = make_ranked_plot(ax = sns.heatmap([p_ab_tp_tn_mean], 
                 xticklabels=[], 
                 yticklabels=['Avg of\nranks\n(actual)'], 
                 robust=True, 
                 annot=True,fmt='.1f', 
                 square=True,
                 cbar=False,
                 cmap=sns.cm.rocket_r,
                 ax=f3_ax3
                                 ),
         title='', xlabel='', ylabel='', dict_colors=dict_sims_colors)
ax_rank_avg.text(letters_space, 0.6,'D', fontsize=25)

# Subplot E
sns.set(font_scale=1.5)
ax_rank_of_sums = make_absolute_plot(ax = sns.heatmap([totalRank_tp_tn], 
                 xticklabels=df_attn_scores['similarities'], 
                 yticklabels=['Final\nrank\n(actual)'], 
                 robust=True, 
                 annot=True,fmt='.1g', 
                 square=True,
                 cbar=False,
                 cmap=sns.cm.rocket_r,
                 ax=f3_ax4
                                 ),
         title='', xlabel='', ylabel='', dict_colors=dict_sims_colors)
ax_rank_of_sums.text(letters_space, 0.6,'E', fontsize=25)

plt.show()
```
